```python
def kdePlot(df, x, y):
    plt.figure(figsize = (15,8))
    left = df.loc[df['left'] == 1]
    stayed = df.loc[df['left'] == 0]
    sns.kdeplot(left[x], left[y], cmap = "Greens", label = "Left")
    sns.kdeplot(stayed[x], stayed[y], cmap = "Blues", label = "Stayed")
kdePlot(df, 'last_evaluation', 'average_montly_hours')
plt.title("Joint Kernel Density Plot of Average Monthly Hours v. Last Evalutation", size = 15)
plt.show()
```

```python
str_list = [] # empty list to contain columns with strings (words)
for colname, colvalue in df.iteritems():
    if type(colvalue[1]) == str:
         str_list.append(colname)
```

```python
str_list = []
numeric_list = []
for column in df:
    if df[column].dtype == object:
        str_list.append(column)
    else:
        numeric_list.append(column)        
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, confusion_matrix, precision_recall_curve
from sklearn.preprocessing import RobustScaler
categorical = pd.get_dummies(data=df[str_list])
numeric = df[numeric_list]
df = pd.concat([numeric, categorical], axis = 1)
```

# What do I do with this and how do I convince my boss

```python
target_name = 'left'
Xu = df.drop('left', axis=1)
robust_scaler = RobustScaler()
X = robust_scaler.fit_transform(Xu)
y=df[target_name]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=123, stratify=y)
```

```python
from sklearn.naive_bayes import GaussianNB
NaiveBayes = GaussianNB()
NaiveBayes.fit(X_train, y_train)
NaiveBayes_model = NaiveBayes.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy of the Naive Bayes Classifier is: " + str(accuracy_score(y_test, NaiveBayes_model)))
```

```python
from sklearn.metrics import confusion_matrix
plt.figure(figsize = (8,8))

mat = confusion_matrix(y_test, NaiveBayes_model)

sns.heatmap(mat, square=True, annot=True, cbar=False, fmt = 'g')
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```python
from sklearn.ensemble import RandomForestClassifier 
rfmodel = RandomForestClassifier()
rfmodel.fit(X_train, y_train)
rfpmodel = rfmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy of the Random Forest Classifier is: " + str(accuracy_score(y_test, rfpmodel)))
```

```python
from sklearn.metrics import confusion_matrix
plt.figure(figsize = (8,8))

mat2 = confusion_matrix(y_test, rfpmodel)

sns.heatmap(mat2, square=True, annot=True, cbar=False, fmt = 'g')
plt.xlabel('predicted value')
plt.ylabel('true value');
```

Business goals:
